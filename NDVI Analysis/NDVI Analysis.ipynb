## Importing the Libraries

```python
import rasterio as rio
import numpy as np
import matplotlib.pyplot as plt
```

## Importing Required Bands

```python
# Defining the image path.
img_path = './data/bangalore.tiff'

# Reading red and nir bands in the RGBN image.
with rio.open(img_path) as src:
    red = src.read(1)
    nir = src.read(4)
```

## Calculating NDVI

```python
# Ignoring floating-point errors - division by zero and invalid floating point operations.
np.seterr(divide='ignore', invalid='ignore')

# Calculating NDVI.
ndvi = (nir.astype(float) - red.astype(float))/(nir + red)
```

## Creating a NDVI Image 

```python
# Saving meta data of the original image in a variable.
meta = src.meta
# Saving the data type of NDVI array. 
ndvi_dtype = ndvi.dtype

# Copying meta data to a new varibale and updating the dtype, band count and driver.
kwargs = meta
kwargs.update(dtype = ndvi_dtype)
kwargs.update(count = 1)
kwargs.update(driver = 'GTiff')

# Saving the NDVI image.
with rio.open('ndvi_bangalore.tiff', 'w', **kwargs) as dst:
    dst.write(ndvi, 1)
```

Normalize class is imported from matplotlib, which normalizes data linearly. A subclass of Normalize is created by the name MidpointNormalize.

```python
# Importing Normalize class from Matplotlib. 
from matplotlib.colors import Normalize

# Creating a subclass for the above imported class.
class MidpointNormalize(Normalize):
    def __init__(self, vmin = None, vmax = None, midpoint = None, clip = False):
        # Assigning attribute.
        self.midpoint = midpoint
        # Using __init__ method from Normalize class
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip = None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y), np.isnan(value))
```

## Interpreting the NDVI Image

```python
# Setting up the dimensions of the figure.
fig = plt.figure(figsize = (10,6))
ax = fig.add_subplot(111)

# Setting up minimum and maximum value
min = np.nanmin(ndvi)
max = np.nanmax(ndvi)

# Setting up the midpoint,
mid=0.1
# Calling MidpointNormalize.
norm = MidpointNormalize(vmin = min, vmax = max, midpoint = mid)
# Assignming a colourmap.
colormap = plt.cm.RdYlGn

# Specifying input data, colourmap, and normalization.
cbar_plot = ax.imshow(ndvi, cmap = colormap, norm = norm)
# Adding a colourbar to the plot.
cbar = fig.colorbar(cbar_plot, orientation = 'vertical', shrink = 1)

# Adding embellishments to the plot. 
ax.axis('off')
ax.set_title('Normalized Difference Vegetation Index (NDVI)', fontsize = 12, fontweight = 'bold')

# Saving NDVI map.
fig.savefig('ndvi_bangalore_colourmap.png', dpi = 200, bbox_inches = 'tight', pad_inches = 0.7)
plt.show()
```
