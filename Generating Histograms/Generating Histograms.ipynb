## Importing Libraries

```python
import rasterio as rio
import matplotlib.pyplot as plt
from rasterio.plot import show
from rasterio.plot import show_hist
import cv2
```

## Importing the Image

```python
img = rio.open('bangalore.tiff')
```

## Plotting the Bands of the Image

```python
plt.figure(figsize = (24, 6))
plt.subplot(1, 4, 1)
plt.imshow(img.read(1), cmap = 'Reds')
plt.subplot(1, 4, 2)
plt.imshow(img.read(2), cmap = 'Greens')
plt.subplot(1, 4, 3)
plt.imshow(img.read(3), cmap = 'Blues')
plt.subplot(1, 4, 4)
plt.imshow(img.read(4), cmap = 'Greys')
```

# Creating a Histogram

```python
plt.figure(figsize = (12, 5))
show_hist(
    img.read([1,2,3]), bins = 51, histtype='stepfilled', title="Histogram", alpha = 0.3, label = ['Red', 'Green', 'Blue'])
```
