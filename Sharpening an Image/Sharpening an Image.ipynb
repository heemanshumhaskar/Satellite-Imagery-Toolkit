## Importing the Libraries|

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Importing the Image

```python
# Reading the Image.
img = cv2.imread('./data/2022-04-28.png')

# Displaying the image.
plt.imshow(img)
```

OpenCV imports an image in BGR format (reverse of RGB), therefore the image slightly looks off from true colour.

## Image Processing

```python
# Creating a kernel which transforms the image.
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])

# Applying the kernel.
img_sharp = cv2.filter2D(src = img, ddepth = -1, kernel = kernel)

# Displaying transformed image.
plt.imshow(img_sharp)
```

## Saving the Image

```python
# Saving the output path string in a variable
output_path = './data/output'

# Checking if the output folder exists. If it does, deleting the older folder and making a new one.
import shutil
if os.path.exists(output_path):
    shutil.rmtree(output_path)
os.mkdir(output_path)

# Saving the image.
cv2.imwrite(os.path.join(output_path, '2022-04-28_sharp.png'), img_sharp)
```
